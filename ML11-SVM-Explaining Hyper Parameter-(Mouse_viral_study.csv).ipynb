## Hyper Parameters

### C

Regularization parameter. The strength of the regularization is **inversely** proportional to C. Must be strictly positive. The penalty is a squared l2 penalty.

Regularizasyon parametresi. Düzenlileştirmenin gücü C ile ters orantılıdır. Kesinlikle pozitif olmalıdır. SVM'de hem lineer olarak ayrılabilir hem de lineer olmayan çözümlerle ilgilenirken hiper parametre olarak geçirilen ceza terimi 'C' olarak ifade edilir. C büyüdükçe margin küçülür ve hatasız ayrıştırma yapmaya doğru gider, overfit durumu olur. yukardaki örnekte olduğu gibi hatasız ayrışma olur buna hard marjin denir. Hard marjin değerine ulaşTıktan sonra C değerini ne kadar büyütürsek büyütelim sınır değişmez. C değerine bağlı olarak margin değişimini aşağıda deneyerek gör. 

C, 0.01, 0.1, 1, 10 olarak aşağıda dene.

*Note: If you are following along with the equations, specifically the value of C as described in ISLR, C in scikit-learn is **inversely** proportional to this value.*

```python
model = SVC(kernel='linear', C=0.01)
model.fit(X, y)
```

```python
plot_svm_boundary(model,X,y)
```

### Kernel

Kernel Function is a method used to take data as input and transform into the required form of processing data. “Kernel” is used due to set of mathematical functions used in Support Vector Machine provides the window to manipulate the data. So, Kernel Function generally transforms the training set of data so that a non-linear decision surface is able to transformed to a linear equation in a higher number of dimension spaces. Basically, It returns the inner product between two points in a standard feature dimension.

[Choosing a Kernel](https://stats.stackexchange.com/questions/18030/how-to-select-kernel-for-svm?rq=1)

#### rbf - [Radial Basis Function](https://en.wikipedia.org/wiki/Radial_basis_function_kernel)


When training an SVM with the Radial Basis Function (RBF) kernel, two parameters must be considered: C and gamma. The parameter C, common to all SVM kernels, trades off misclassification of training examples against simplicity of the decision surface. A low C makes the decision surface smooth, while a high C aims at classifying all training examples correctly. gamma defines how much influence a single training example has. The larger gamma is, the closer other examples must be to be affected.

kernel boyut arttırmayı sağlar. ('linear', 'poly', 'rbf', 'sigmoid', 'precomputed') gibi yöntemleri kullanır. "rbf" tercih edeceğimiz yöntem. "linear" dışında kernel kullandığımızda C ile birlikte gamma da devreye girer. 

Düşük gama değerleri, daha fazla noktanın birlikte gruplandırılmasıyla sonuçlanan büyük bir benzerlik yarıçapını gösterir yani daha geniş bir daire çizer. Yüksek gama değerleri için aynı grupta (veya sınıfta) sayılabilmesi için noktaların birbirine çok yakın olması gerekir ve çok dar bir daire çizer. Bu nedenle, çok büyük gama değerlerine sahip modeller overfit eğilimindedir.

gamma nın büyümeside overfite götürür. büyük gamma değeri C değerinide bastırır etkisiz hale getirir. gamma dominanttır. 

linear kernel için sadece C parametresini optimize etmemiz gerekir. RBF kernel kullanmak istiyorsak, hem C hem de gamma parametresinin aynı anda optimize edilmesi gerekir. Gama büyükse, C'nin etkisi ihmal edilebilir hale gelir etkisizleşir. Gamma küçükse, C modeli doğrusal bir modeli etkilediği gibi etkiler. 

aşağıda c ve gamma değerleri ile oynayarak durumu gör

```python
model = SVC(kernel='rbf', C=10, gamma=0.5)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

### gamma

gamma : {'scale', 'auto'} or float, default='scale'
    Kernel coefficient for 'rbf', 'poly' and 'sigmoid'.

    - if ``gamma='scale'`` (default) is passed then it uses
      1 / (n_features * X.var()) as value of gamma,
    - if 'auto', uses 1 / n_features.

```python
model = SVC(kernel='rbf', C=1, gamma=10)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

The sigmoid kernel was quite popular for support vector machines due to its origin
from neural networks. Although it is known that the kernel matrix may not be positive
semi-definite (PSD), other properties are not fully studied. In this paper, we discuss
such non-PSD kernels through the viewpoint of separability. Results help to validate
the possible use of non-PSD kernels. One example shows that the sigmoid kernel matrix
is conditionally positive definite (CPD) in certain parameters and thus are valid kernels
there. However, we also explain that the sigmoid kernel is not better than the RBF kernel
in general

```python
model = SVC(kernel='sigmoid')
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

#### Degree (poly kernels only)

Degree of the polynomial kernel function ('poly').
Ignored by all other kernels.

```python
model = SVC(kernel='poly', C=1, degree=3)
model.fit(X, y)
plot_svm_boundary(model,X,y)
```

## Grid Search

Keep in mind, for this simple example, we saw the classes were easily separated, which means each variation of model could easily get 100% accuracy, meaning a grid search is "useless".

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.01,0.1,1,10,1000],
              'kernel':['linear','rbf','sigmoid','poly'],
              'gamma':["scale", "auto"],
              'degree':[1,2]}
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```

```python
grid.cv_results_
```

```python
grid.best_score_
```

```python
grid.best_params_
```
